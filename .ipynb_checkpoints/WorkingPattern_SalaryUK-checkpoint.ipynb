#### Business Question 2: Does people work from home earn more or less than people work on-site in UK?
#### In this notebook, We are curious about seeing if different working pattern leads to a difference in developer salary in UK. 
#### To dig into this question, we use columns 'RemoteWork', 'CompTotal', 'Employment' and 'Country'

####  Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

####  Load Data

```python
data = pd.read_csv('./survey_results_public.csv')
data.head()
```

#### Data Exploration - Check what each feature means

```python
data['Country'].unique()
```

```python
data['Employment'].unique()
```

```python
# find the corresponding column
data['RemoteWork'].value_counts()
```

```python
data[['CompTotal']].describe()
```

####  Clean Data 
- There is no missing value in 'Employment' column and very few in 'Country' column, so we decide to drop these as it's expected not to impact the analysis that much.
- While for 'RemoteWork' column, it contains 3 categories representing remote working, on-site working and hybrid working, which has its own distribution. In consideration of the percentage (over 16%), replacing the missing value with any of three would change the distribution and impact our analysis, so we choose to drop them as well.
- In Salary column, there is a large portion of missing values. From the statistic result of this column, we found out there are many outliers, e.g., minimum value of 0, maximum value over 1 million. All of these would impact the statistic values like mean and medium in a significant ways, with which we usually replace the missing value. We are very cautious to replace such a big portion with one value that may have been affected by outliers, as it may cause disruption in the proportion of the data. So we would rather decide to drop missing values than populate them improperly.

```python
cols = ['Country', 'RemoteWork', 'Employment', 'CompTotal']
df = data[cols]
```

```python
# null vaues
df.isna().sum()/data.shape[0]
```

```python
df = df.dropna()
df
```

####  Filter on Data - to only focus on UK full-time developers

```python
# filter on certain conditions
cond_country= (df.Country == 'United Kingdom of Great Britain and Northern Ireland')
cond_emply = (df.Employment.str.contains('Employed, full-time'))

dt = df[cond_country & cond_emply]
dt
```

#### Visualize Result

```python
result1 = dt.groupby('RemoteWork')['CompTotal'].agg(
    count='count',
    mean='mean',
    min='min',
    median='median',
    max='max'
).reset_index()

result1
```

##### The result looks not quite convincing as salary is not supposed to be 0 (unless developers are doing volunteering work, which should be excluded in further analysis)

```python
# Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='RemoteWork', y='CompTotal', data=dt, palette='coolwarm')
plt.title('Salary Distribution by Work Pattern')
plt.show()
```

```python
# Check for extreme values in the 'CompTotal' column
dt['CompTotal'].describe()

# Identify rows with extremely high salaries
high_salary = dt[dt['CompTotal'] >= 1e6]
high_salary
```

```python
# Remove rows where 'CompTotal' is greater than 1 million, set a min threshold 100.0
dt_filtered = dt[(dt['CompTotal'] > 100.0) & (dt['CompTotal'] < 1e6)]

# Create a new boxplot with the filtered data
plt.figure(figsize=(10, 6))
sns.boxplot(x='RemoteWork', y='CompTotal', data=dt_filtered, palette='coolwarm')
plt.title('Salary Distribution by Work Pattern (Filtered)')
plt.show()
```

```python
result2 = dt_filtered.groupby('RemoteWork')['CompTotal'].agg(
    count='count',
    mean='mean',
    min='min',
    median='median',
    max='max'
).reset_index()

result2
```

#### Conclusion - The box plot and statistics should be analyzed together for the conclusion. 
#### From the statistics, we firstly notice that there is some imbalance among three working patterns in UK, most people work remotely or in a hybrid way (several days remotely and some time on site), less than 200 people work fully on-site. The box plot shows the salary between hybrid working and remote working is roughly similar, while for in-person working groups the income is slightly lower than the other two. Developers may benefit from the result on a wiser career decision as well as employers make better offers to retain talents.
